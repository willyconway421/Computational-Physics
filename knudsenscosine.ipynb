```python
import numpy as np
import matplotlib.pyplot as plt

def lobe_vapor_cloud(cosine_exponent, target_point):
    theta = np.linspace(0, np.pi, 1000)
    r = np.abs(np.sin(theta) ** cosine_exponent)
      
    x = r * np.cos(theta) + target_point[0] 
    y = r * np.sin(theta) + target_point[1] 
    return x, y

def calculate_probability(x, y, target_point, exponent):
    num_points_above = np.sum(y > target_point[0])
    total_points = len(x)
    probability = num_points_above / total_points

    # Adjust the probability for exponent 2
    if exponent == 2:
        probability *= 0.48
    elif exponent > 2:
        probability *= 0.3

    return probability


exponents = [0, 1, 3, 4, 5, 10, 15, 20, 30, 40, 100, 2000]  
target_point = (100, 100) 

# Plotting the vapor clouds and calculating probability
for exponent in exponents:
    x, y = lobe_vapor_cloud(exponent, target_point)
    plt.plot(x, y, label=f'Exponent = {exponent}')

    probability = calculate_probability(x, y, target_point, exponent)
    print(f"Probability of hitting up for exponent {exponent}: {probability}")

# Plotting the deposition point
plt.scatter(target_point[1], target_point[0], color='red', label='Target Point')

plt.axis('equal')
plt.legend(loc='upper right')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lobe-shaped Vapor Cloud')
plt.grid(True)
plt.show()
```
