```python
# Define the function g(x)
def g(x):
    return 2*x**4 - 6*x**3 - 14*x - 8

# Define forward difference approximation for the first derivative
def forward_diff(f, x, h):
    frd_diff = (f(x + h) - f(x)) / h
    return frd_diff

# Define backward difference approximation for the first derivative
def back_diff(f, x, h):
    bck_diff = (f(x) - f(x - h)) / h
    return bck_diff

# Define central difference approximation for the first derivative
def central_diff(f, x, h):
    cntrl_diff = (f(x + h) - f(x - h)) / (2 * h)
    return cntrl_diff

# Define initial values for x and h
x1 = 0
h1 = 0.01

# Calculate first derivatives using different difference schemes
f1 = forward_diff(g, x1, h1)
b1 = back_diff(g, x1, h1)
c1 = central_diff(g, x1, h1)

# Calculate second derivatives using different difference schemes
f2 = (forward_diff(g, x1 + h1, h1) - forward_diff(g, x1, h1)) / h1
b2 = (back_diff(g, x1, h1) - back_diff(g, x1 - h1, h1)) / h1
c2 = (central_diff(g, x1 + h1, h1) - central_diff(g, x1 - h1, h1)) / (2 * h1)

# Print the results
print("The first derivative using forward, backward, and central differences are", round(f1, 3), round(b1, 3), "and", round(c1, 3), "respectively.")
print("The second derivative using forward, backward, and central differences are", round(f2, 3), round(b2, 3), "and", round(c2, 3), "respectively.")
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81  # Acceleration due to gravity (m/s^2)
m = 68    # Mass of the object (kg)
v = 12.5  # Terminal velocity (m/s)
e = math.e  # Euler's number

# Function to compute velocity at time t
def velocity(t1):
    vel = (g * m / v) * (1 - e**(-v * t1 / m))
    return vel

# Function to compute distance travelled at time t
def dist(t2):
    dist = (g * m / v) * t2 + (g * m**2 / v**2) * e**(-v * t2 / m)
    return dist

# Initialize time and distance
t = 0
x = 0
time = []
distance = []

# Simulate the motion and compute distance travelled at each time step
while t <= 16:
    dx = velocity(t) * 0.001  # Time step of 0.001 seconds
    x = x + dx
    time.append(t)
    distance.append(x)
    t += 0.001  # Increment time by 0.001 seconds

# Print distance travelled computationally
print("Distance travelled by the parachute in 16 seconds is", round(x, 3), "m when solved computationally")

# Calculate distance travelled analytically using the dist function
r = dist(16) - dist(0)
print("Distance travelled by the parachute in 16 seconds is", round(r, 3), "m when solved analytically")

# Plot the distance travelled over time
plt.plot(time, distance)
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.title("Distance vs Time for Parachute Descent")

plt.show()
```
