```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_system(N, nsteps):
    nl = N // 2  # Number of particles on the left side initially
    nr = N - nl  # Number of particles on the right side initially

    # Arrays to store the number of particles on each side at each time step
    nl_values = np.zeros(nsteps)
    nr_values = np.zeros(nsteps)

    for step in range(nsteps):
        # Probability for a move to the right during a time step ∆t
        move_probability = nl / N

        # Generate a random number
        x = np.random.rand()

        if x <= move_probability:
            nl -= 1
            nr += 1
        else:
            nl += 1
            nr -= 1

        # Store the values for this time step
        nl_values[step] = nl
        nr_values[step] = nr

    return nl_values, nr_values

# Parameters
N = 30
nsteps = 1000 * N

# Simulation
nl_values, nr_values = simulate_system(N, nsteps)

# Plotting
time = np.arange(nsteps)
plt.plot(time, nl_values, label='Left side')
plt.plot(time, nr_values, label='Right side')
plt.xlabel('Time steps')
plt.ylabel('Number of particles')
plt.title('Number of particles in each half over time')
plt.legend()
plt.show()
```

```python
# Equilibrium determination
equilibrium_time = np.where(np.abs(nl_values - nr_values) <= 1)[0][0]
print("Equilibrium reached at time step:", equilibrium_time)

# Plot distribution during equilibrium
plt.hist(nl_values[equilibrium_time:], bins=range(N+1), density=True, alpha=0.7)
plt.xlabel('Number of particles in the left half')
plt.ylabel('Probability')
plt.title('Distribution of particles in the left half at equilibrium')
plt.show()
```
