```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate the particle movement
def simulate(N, nsteps):
    nl = N  # Initial number of particles on the left side
    nr = 0  # Initial number of particles on the right side

    left_counts = [nl]  # List to store number of particles on the left side
    right_counts = [nr]  # List to store number of particles on the right side

    for _ in range(nsteps):
        # Probability for a particle to move to the right
        prob_move_right = nl / N
        
        # Generate a random number between 0 and 1
        x = np.random.rand()

        # Move particle from left to right
        if x <= prob_move_right:
            nl -= 1
            nr += 1
        # Move particle from right to left
        else:
            nl += 1
            nr -= 1

        # Append the counts to the lists
        left_counts.append(nl)
        right_counts.append(nr)

    return left_counts, right_counts

# Parameters
N = 50
nsteps = 1000 * N

# Simulate the system
left_counts, right_counts = simulate(N, nsteps)

# Time array
time = np.arange(nsteps + 1)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, left_counts, label='Left Half')
plt.plot(time, right_counts, label='Right Half')
plt.xlabel('Time')
plt.ylabel('Number of Particles')
plt.title('Particle Distribution Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate the particle movement
def simulate(N, nsteps):
    nl = N  # Initial number of particles on the left side
    nr = 0  # Initial number of particles on the right side

    left_counts = [nl]  # List to store number of particles on the left side
    right_counts = [nr]  # List to store number of particles on the right side

    for _ in range(nsteps):
        # Probability for a particle to move to the right
        prob_move_right = nl / N
        
        # Generate a random number between 0 and 1
        x = np.random.rand()

        # Move particle from left to right
        if x <= prob_move_right:
            nl -= 1
            nr += 1
        # Move particle from right to left
        else:
            nl += 1
            nr -= 1

        # Append the counts to the lists
        left_counts.append(nl)
        right_counts.append(nr)

    return left_counts, right_counts

# Parameters
N = 30
nsteps = 1000 * N

# Simulate the system
left_counts, right_counts = simulate(N, nsteps)

# Time array
time = np.arange(nsteps + 1)

# Find equilibrium point
equilibrium_time = np.argmax(np.abs(np.diff(left_counts[-1000:])))  # Check the last 1000 time steps

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, left_counts, label='Left Half')
plt.plot(time, right_counts, label='Right Half')
plt.axvline(x=equilibrium_time, color='red', linestyle='--', label='Equilibrium')
plt.xlabel('Time')
plt.ylabel('Number of Particles')
plt.title('Particle Distribution Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Plot distribution of particles in the left half during equilibrium
plt.figure(figsize=(8, 6))
plt.hist(left_counts[equilibrium_time:], bins=range(N+1), density=True, alpha=0.7, color='blue')
plt.xlabel('Number of Particles in Left Half')
plt.ylabel('Probability')
plt.title('Distribution of Particles in Left Half at Equilibrium')
plt.grid(True)
plt.show()

print(f"System reaches equilibrium after {equilibrium_time} time steps.")
```
